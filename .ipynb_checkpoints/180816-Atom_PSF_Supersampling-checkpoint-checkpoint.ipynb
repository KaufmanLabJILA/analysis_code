```python
def radial_profile(data, center):
    y, x = np.indices((data.shape))
    r = np.sqrt((x - center[0])**2 + (y - center[1])**2)
    r = r.astype(np.int)

    tbin = np.bincount(r.ravel(), data.ravel())
    nr = np.bincount(r.ravel())
    radialprofile = tbin / nr
    return radialprofile
```

```python
def get_losses(n):
    flat = np.array(roisums[n]).flatten('F')/countsperphoton
    oddcounts = flat[1::2]
    evencounts = flat[::2]
    odds = np.array([int(x) for x in oddcounts>oddthresh])
    evens = np.array([int(x) for x in evencounts>eventhresh])
    ne, bse, ps = plt.hist(evencounts,50, alpha = 1, histtype = 'step')
    #plt.show()
    n, bs, ps = plt.hist(oddcounts,50, alpha = 1, histtype = 'step')
    #plt.axis([-10, 50, 0, 150])
    #plt.show()
    #plt.plot(bins, 'ko')
    #plt.axis([0, 100, 0, 2])
    plt.show()
    sums = odds + evens
    diffs = evens - odds
    atoms = len(sums[sums > 1])
    voids = len(sums[sums < 1])
    losses = len(sums[diffs == 1])
    return (losses, losses+atoms, )

losses = np.array([get_losses(n) for n in range(len(keysort))])
losses, tots, missing = zip(*losses)
losses = np.array(list(losses))
tots = np.array(list(tots))

```
